```python
#importing required modules
import datetime
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import butter,filtfilt,freqz
from sklearn import *
from sklearn.metrics import f1_score
import os
```

```python
file='./processed_data/DetailVolDownsampled.csv'
```

```python
detail_df=pd.read_csv(file)
```

```python
batch_size=len(detail_df['Auxiliary channel TU1 U(V)'])
num_batches=10
res=1000
fs = 1       # sample rate, 10kHz
nyq = 0.5 * fs  # Nyquist Frequency
cutoff_freq_sweep = range(250,4750,50) # Sweeping from 250 to 4750 Hz for SNR measurement
lpf_cutoff = 600
```

```python
plt.figure(figsize=(20,5))
plt.plot(range(0,detail_df.shape[0],res),detail_df['Auxiliary channel TU1 U(V)'][0::res])
for i in range(num_batches+1): plt.plot([i*batch_size,i*batch_size],[-5,12.5],'r')
for j in range(num_batches): plt.text(j*batch_size+200000,num_batches,str(j+1),size=20)
plt.xlabel('Row',size=16); plt.ylabel('Signal',size=16); 
plt.title('Training Data Signal - 10 batches',size=20)
plt.show()
```
